```python
from selenium.webdriver.chrome.options import Options
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import NoSuchElementException
from selenium.webdriver.support.ui import WebDriverWait
```

```python
url_2 = "https://www.imdb.com/search/title/?title_type=feature,tv_series&num_votes=25000,&genres=family&sort=user_rating,desc"

options = Options()
# Set language explicitly in Chrome
options.add_argument('--lang=en-US')
# Set accept-language header
options.add_argument('--accept-language=en-US,en;q=0.9')
# Add user agent to ensure English content
options.add_argument('--user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36')

# Force Chrome to launch in English
options.add_experimental_option('prefs', {
    'intl.accept_languages': 'en,en_US',
    'translate': {'enabled': 'true'},
    'translate_whitelists': {'de': 'en'}
})

driver = webdriver.Chrome(options=options)

# Set language cookie before loading the page
driver.get("https://www.imdb.com")
driver.add_cookie({'name': 'language', 'value': 'en-US'})

# Now load the actual page
driver.get(url_2)
time.sleep(5)

while True:
    try:
        # Find the button
        view_more_button = driver.find_element(By.CSS_SELECTOR, '.ipc-see-more__button')
        
        # Scroll the button into view
        driver.execute_script("arguments[0].scrollIntoView(true);", view_more_button)
        # Add a small delay after scrolling
        time.sleep(1)
        
        # Add additional scroll to ensure button is not covered by any floating elements
        driver.execute_script("window.scrollBy(0, -100);")
        time.sleep(1)
        
        # Wait for button to be clickable
        WebDriverWait(driver, 10).until(
            EC.element_to_be_clickable((By.CSS_SELECTOR, '.ipc-see-more__button'))
        )
        
        # Click the button
        view_more_button.click()
        time.sleep(2)
        
    except NoSuchElementException:
        print("No more 'View More' button found.")
        break
    except Exception as e:
        print(f"An error occurred: {e}")
        break

html = driver.page_source
soup = BeautifulSoup(html, 'html.parser')

# Don't forget to quit the driver
driver.quit()
```

```python
titles = []

for index, row in ibdm_df.iterrows():
    titles.append(row['api'])
```

```python
#Now I need to write a loop for the API call
import json
```

```python
def getMovieData():
    result = []
    for title in range(len(titles)):
        url = 'http://www.omdbapi.com/?t={}&apikey=3cb24584'.format(titles[title])
        #print(url)
        req = requests.get(url)
        data = req.json()
        content = {
            "Title": data['Title'],
            "Genre": data['Genre'],
            "Release Date": data['Released'],
            "Plot": data['Plot']}
        result.append(content)
    return result

data = getMovieData()
```

```python
movie_df = pd.DataFrame(data)
```

```python
# Next, I need to separate the year in a separate column
# Then I need to create a column for every single Genre, so they can be counted separately 
```

```python
import re
```

```python
def split(year):
    return re.findall('(\d\d\d\d)', year)

movie_df["Release Date"].apply(split)
```

```python
movie_df["Year"] = movie_df["Release Date"].apply(split)
```

```python
# split the Genre column into three columns
movie_df[['First Genre', 'Second Genre', 'Third Genre']] = movie_df['Genre'].str.split(',', expand=True)
```

```python
movie_df.head()
```

```python
#removing the brackets from the year column

movie_df['Year'] = movie_df['Year'].str[0]
```

```python
movie_df.head()
```

```python
#Now I am going to analyze how many times each Genre was represented in a certain year

movie_df[['First Genre','Second Genre','Third Genre','Year']].groupby('Year').value_counts()
```

```python
movie_df[['First Genre','Year']].groupby('First Genre').value_counts()
```

```python
#Melting it all together 

melted_df = movie_df.melt(id_vars=['Year'], 
                    value_vars=['First Genre', 'Second Genre', 'Third Genre'],
                    var_name='genre_column', 
                    value_name='genre',
                    ignore_index=False)

melted_df['genre'] = melted_df['genre'].str.strip().str.title()

# Group by year and genre, count occurrences
genre_counts = melted_df.groupby(['Year', 'genre']).size().unstack(fill_value=0)
```

```python
melted_df.head()
```

```python
genre_counts
```

```python
#Counting which Genres are most popular over time

genre_counts.sum().sort_values(ascending=False)
```

```python
genre_counts = genre_counts.reset_index()
```

```python
# Now I need the most popular genres for each year
genre_counts['Total'] = genre_counts.sum(axis=1)

yearly_totals_sorted = genre_counts.sort_values('Total', ascending=False)
```

```python
#This shows me in what year there was the biggest number of genres in total

yearly_totals_sorted
```

```python
#Now checking how many genres were used each year 

unique_years = genre_counts["Year"].unique()

# Initialize a dictionary to store genre counts by year
genre_counts_by_year = {}

for year in unique_years:
    print(f"\nProcessing Year: {year}")
    year_data = genre_counts[genre_counts["Year"] == year]
    genre_data = year_data.drop(columns=["Year"])
    genre_count = (genre_data > 0).any(axis=0).sum()
    genre_counts_by_year[year] = genre_count
    print(f"Number of genres used in {year}: {genre_count}")
```
